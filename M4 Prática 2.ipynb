```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
```

```python
# Supondo que dados_acerto esteja definido como em ipython
dados_acerto = pd.DataFrame(np.random.randint(0, 25, (100, 7)), columns=['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química'])
n_questões = [24, 18, 8, 8, 8, 12, 12]

# Calcule pontuações percentuais para cada assunto
dados_acerto_percentual = dados_acerto / n_questões * 100

# Atribuir as colunas de dados_acerto_percentual às variáveis ​​utilizadas em pd.concat
pc_mat = dados_acerto_percentual['Matemática']
pc_por = dados_acerto_percentual['Português']
pc_geo = dados_acerto_percentual['Geografia']
pc_ing = dados_acerto_percentual['Inglês']
pc_his = dados_acerto_percentual['História']
pc_fis = dados_acerto_percentual['Física']
pc_qui = dados_acerto_percentual['Química']
```

```python
# Agora você pode concatenar a série em um DataFrame
df = pd.concat([pc_mat, pc_por, pc_geo, pc_ing, pc_his, pc_fis, pc_qui], axis=1)
df.columns = ['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química']
df.head()
```

```python
# Substituir NaN pela média de cada coluna (apenas para colunas numéricas)
for column in df.select_dtypes(include=np.number).columns: # Iterate through only numeric columns
    df[column].fillna(df[column].mean(), inplace=True)
# Exibir o DataFrame resultante
print(df)
```

```python
# Obtenha os nomes das colunas atuais
current_columns = df.columns.tolist()

# Substitua os primeiros 9 nomes de colunas pelos novos nomes, garantindo 'APGAR5' foi renomeado
new_columns = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
df = df.rename(columns={current_columns[i]: new_columns[i] for i in range(len(new_columns))})

# Ver o resultado
print(df)
```

```python
# Verificar colunas duplicadas
duplicatas = df.columns[df.columns.duplicated()].tolist()
print("Colunas duplicadas:", duplicatas)
```

```python
# Renomear colunas duplicadas
df.columns = [f"{col}_{i}" if col in duplicatas else col for i, col in enumerate(df.columns)]
```

```python
print(df)
```

```python
# Definindo as condições para classificação
condicoes = [
    (df['APGAR5_8'] >= 8) & (df['APGAR5_8'] <= 10),  # Normal
    (df['APGAR5_8'] >= 6) & (df['APGAR5_8'] <= 7),   # Asfixia Leve
    (df['APGAR5_8'] >= 4) & (df['APGAR5_8'] <= 5),   # Asfixia Moderada
    (df['APGAR5_8'] >= 0) & (df['APGAR5_8'] <= 3)    # Asfixia Severa
]

# Definir os rótulos correspondentes
rotulos = ['normal', 'asfixia leve', 'asfixia moderada', 'asfixia severa']

# Criar a nova coluna de classificação
df['classificacao_asfixia'] = np.select(condicoes, rotulos, default=np.nan)

# Calcular as frequências das categorias
frequencias = df['classificacao_asfixia'].value_counts()

# Exibir o resultado
print(frequencias)
```

```python
# Renomear as colunas para snake case
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Exibir o DataFrame com as colunas renomeadas
print("\nDataFrame com colunas renomeadas:")
print(df)
```

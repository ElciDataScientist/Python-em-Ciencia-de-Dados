```python
data = {
    'mes_00': [0, 1, 0],
    'mes_01': [0, 0, 1],
    'mes_02': [0, 1, 0],
    'mes_03': [0, 0, 0],
    # ... até 'mes_24'
}

df1 = pd.DataFrame(data)

# Vamos considerar que um valor maior que 0 indica *default*
# Criamos a coluna 'indicador_default' usando `.isin()`
df1['indicador_default'] = df1[['mes_00', 'mes_01', 'mes_02', 'mes_03']].isin([1]).any(axis=1)

print(df1)
```

```python
# Criar uma nova coluna que verifica se o cliente teve pelo menos um episódio de default
df1['teve_default'] = (df1[['mes_00', 'mes_01', 'mes_02', 'mes_03']].sum(axis=1) > 0)

print(df1)
```

```python
df1 = pd.DataFrame({
    'cliente_id': [1, 2, 3, 4, 5],
    'proponente': ['sim', 'sim', 'sim', 'sim', 'sim']
})

# Base de pagamentos com clientes que adquiriram o cartão
pagamentos = pd.DataFrame({
    'cliente_id': [1, 3, 5],
    'pagamento': ['ok', 'ok', 'ok']
})

# Realizando o merge para selecionar apenas clientes que estão nas duas bases
clientes_com_cartao = df1.merge(pagamentos, on='cliente_id', how='inner')

# Marcando os clientes que se tornaram tomadores (os que adquiriram o cartão)
clientes_com_cartao['tomador'] = True

# Exibindo o resultado
print(clientes_com_cartao)
```

```python
df1 = pd.DataFrame({
    'cliente_id': [1, 2, 3, 4, 5],
    'proponente': ['sim', 'sim', 'sim', 'sim', 'sim']
})

# Base de pagamentos com clientes que adquiriram o cartão e o indicador de default
pagamentos = pd.DataFrame({
    'cliente_id': [1, 3, 5],
    'teve_default': [True, False, True]
})

# Realizando o merge para juntar as informações da base de propostas com o indicador de default
df1_com_default = df1.merge(pagamentos, on='cliente_id', how='inner')

# Exibindo o resultado
print(df1_com_default)
```

```python
# Supondo que df1_com_default é o DataFrame resultante do merge com o indicador de default
contagem_default = df1_com_default['teve_default'].value_counts()

# Exibindo a contagem
print(contagem_default)
```

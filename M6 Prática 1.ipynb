```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dados = pd.DataFrame({
    'APGAR1': [8, 9, 7, 9, 8],
    'APGAR5': [9, 9, 8, 9, 9],
    'SEXO': ['M', 'F', 'M', 'F', 'M'],
    'GESTACAO': [38, 40, 39, 41, 37],
    'ESTCIVMAE': ['Casada', 'Solteira', 'Casada', 'Divorciada', 'Casada']
})
```

**Gráfico de barras para APGAR1**

```python
plt.figure(figsize=(8, 6))
dados['APGAR1'].value_counts().sort_index().plot(kind='bar', color='blue')
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.show()
```

**Gráfico de barras para GESTACÃO**

```python
dados = pd.DataFrame({
    'CONSULTAS': ['Mensal', 'Trimestral', 'Mensal', 'Mensal', 'Trimestral'],
    'APGAR5': [9, 8, 9, 7, 8]
})
```

**MÉDIA**

```python
# Calculando a média de APGAR5 por CONSULTAS
media_apgar5_por_consultas = dados.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Renomeando a coluna para maior clareza
media_apgar5_por_consultas.columns = ['CONSULTAS', 'MEDIA_APGAR5']

# Exibindo o resultado
print(media_apgar5_por_consultas)
```

**Gráfico de Barras**

```python
plt.figure(figsize=(8, 6))
plt.bar(media_apgar5_por_consultas['CONSULTAS'], media_apgar5_por_consultas['APGAR5'], color='blue')
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')
plt.show()
```

```python
dados = pd.DataFrame({
    'SEXO': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'GESTACAO': [38, 40, 39, 41, 37, 38, 40, 41, 38, 39],
    'ESTCIVMAE': ['Casada', 'Solteira', 'Casada', 'Divorciada', 'Casada', 'Solteira', 'Casada', 'Divorciada', 'Casada', 'Solteira'],
    'CONSULTAS': ['Mensal', 'Trimestral', 'Mensal', 'Mensal', 'Trimestral', 'Mensal', 'Trimestral', 'Trimestral', 'Mensal', 'Trimestral']
})
```

**Graficos 2x2**

```python
# Exemplo de dados para os gráficos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Criando a figura com subplots (2 linhas, 2 colunas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Ajustando o gráfico em cada eixo
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Seno')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosseno')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Tangente')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Exponencial Decrescente')

# Ajuste para melhorar o espaçamento
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

**Gráfico de barras para SEXO**

```python
# Criando a figura com subplots (2 linhas, 2 colunas)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Gráfico 1: Frequência de SEXO
sns.countplot(data=dados, x='SEXO', ax=axes[0, 0], palette='Set1')
axes[0, 0].set_title('Frequência de SEXO')

# Outros gráficos podem ser adicionados aos eixos restantes (como placeholders)
# Por exemplo, podemos adicionar mais gráficos em cada um dos outros eixos:
sns.countplot(data=dados, x='SEXO', ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Gráfico 2')

sns.countplot(data=dados, x='SEXO', ax=axes[1, 0], palette='Set3')
axes[1, 0].set_title('Gráfico 3')

sns.countplot(data=dados, x='SEXO', ax=axes[1, 1], palette='Set1')
axes[1, 1].set_title('Gráfico 4')

# Ajuste para melhorar o espaçamento
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

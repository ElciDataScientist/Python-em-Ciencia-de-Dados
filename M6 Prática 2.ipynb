```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

```python
peso_do_bebe = [random.randint(2500, 4000) for _ in range(100)]
```

```python
dados = pd.DataFrame({'Peso do Bebê (g)': peso_do_bebe})
```

**Boxplot usando Matplotlib**

```python
plt.figure(figsize=(8, 6))
plt.boxplot(dados['Peso do Bebê (g)'], vert=False)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.show()
```

**Gráfico**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dados, x='Peso do Bebê (g)')
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.show()
```

```python
idade_da_mae = [random.randint(18, 45) for _ in range(200)]
```

```python
dados = pd.DataFrame({'Idade da Mãe': idade_da_mae})
```

**Histograma usando Matplotlib**

```python
plt.figure(figsize=(8, 6))
plt.hist(dados['Idade da Mãe'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

**Histograma usando Seaborn**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x='Idade da Mãe', bins=15, color='lightcoral', kde=True)
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(peso_do_bebe, bins=15, color='skyblue', edgecolor='black')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('Frequência')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

```python
peso_do_bebe = [random.randint(2500, 4000) for _ in range(100)]
```

```python
dados = pd.DataFrame({'Peso do BebÃª (g)': peso_do_bebe})
```

**Boxplot usando Matplotlib**

```python
plt.figure(figsize=(8, 6))
plt.boxplot(dados['Peso do BebÃª (g)'], vert=False)
plt.title('Boxplot do Peso do BebÃª')
plt.xlabel('Peso do BebÃª')
plt.show()
```

**GrÃ¡fico**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dados, x='Peso do BebÃª (g)')
plt.title('Boxplot do Peso do BebÃª')
plt.xlabel('Peso do BebÃª')
plt.show()
```

```python
idade_da_mae = [random.randint(18, 45) for _ in range(200)]
```

```python
dados = pd.DataFrame({'Idade da MÃ£e': idade_da_mae})
```

**Histograma usando Matplotlib**

```python
plt.figure(figsize=(8, 6))
plt.hist(dados['Idade da MÃ£e'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da MÃ£e')
plt.xlabel('Idade da MÃ£e')
plt.ylabel('FrequÃªncia')
plt.show()
```

**Histograma usando Seaborn**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x='Idade da MÃ£e', bins=15, color='lightcoral', kde=True)
plt.title('Histograma da Idade da MÃ£e')
plt.xlabel('Idade da MÃ£e')
plt.ylabel('FrequÃªncia')
plt.show()
```
